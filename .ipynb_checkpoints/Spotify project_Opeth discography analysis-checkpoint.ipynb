# Cleaning the Data

Looks like we have a few issues, here, due to Opeth being included in some Greatest hits albums and also having many “special edition” versions of albums. Let’s drop these albums and create a clean list of album ids:


```python
album_names = []
albums = []

for album in opeth_albums['items']:
    album_name = album['name']
    album_id = album['id']
    # Clean out the duplicates
    if album_name[:3] not in album_names and "Metal" not in album_name and "Rock" not in album_name and "Live" not in album_name and "Song" not in album_name and "Sega" not in album_name:
        album_names.append(album_name[:3] )
        albums.append((album_name,album_id))
```

```python
albums
```

```python
final_album_list = albums
```

```python
final_album_list
```

# Songs from Opeth’s Albums

```python
opeth_song_ids = []

for album_name, album_id in final_album_list:
    tracks = sp.album_tracks(album_id)['items']
    for track in tracks:
        track_name = track['name']
        track_id = track['id']
        opeth_song_ids.append((album_name,album_id,track_name,track_id))
```

```python
opeth_song_ids[0]
```

```python
len(opeth_song_ids)
```

# Song Features

Spotify has machine generated features for songs, let’s grab them with the API.

```python
# Test out one song:
features = sp.audio_features(tracks=['55kTYNyOQQr8rO5pTt1qou'])
```

```python
features
```

```python
pd.Series(features[0])
```

```python
# Getting features for all Opeth songs
df = pd.DataFrame()

for album_name,album_id,track_name, track_id in opeth_song_ids:
    features = sp.audio_features(tracks=[track_id])[0]
    features['album_name'] = album_name
    features['album_id'] = album_id
    features['track_name'] = track_name
    ser = pd.Series(features)
    df = pd.concat([df,ser],axis=1)
```

```python
# Tranpose to fix concat issue due to axis=1
df = df.transpose().set_index('id')
df.head()[['danceability', 'energy', 'key', 'acousticness','album_name','track_name']]
```

# Calculating Opeth’s Songwriting Range

Obviously there is a lot of variance in Opeth’s songwriting across albums, but what songs are the most different? We can calculate the euclidean distance between all the songs and then find the largest distance.

```python
distances = scipy.spatial.distance.cdist(data_features, data_features, metric='euclidean')
```

Here we can see a heatmap of the distance between each song. Obviously, each song will have 0 distance between itself (which is the dark diagonal line you see). But what about this crazy range we see around the 77-80 id mark?

```python
plt.figure(dpi=200)
sns.heatmap(distances)
```

```python
df[distances==distances.max()][['album_name','track_name']]
```

Looks like its “Black Rose Immortal” from Morningrise and “Persephone” from the album Sorceress. So what’s the next step? They do differ in the song lengths(typical Opeth!). Persephone is shorter at 2 mins, purely acoustic and instrumental, unlike Black Rose Immortal(20 mins long!).

Is distance a major feature? Let’s try dropping some data features and trying out this analysis again.
